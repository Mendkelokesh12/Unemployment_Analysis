```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataSets from a CSV file
df=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
# Display the first few rows of the DataFrame
df.head()
```

```python
# Display the last few rows of the DataFrame
df.tail()
```

```python
# Display a random sample of rows from the DataFrame
df.sample()
```

```python
# Check for missing values in the DataFrame
print(df.isnull().sum())
```

```python
# drop on the rows missing values
df=df.dropna()
df.count()
```

```python
# checking the null is remaning or not
print(df.isnull().sum())
```

```python
# Display summary statistics of the DataFrame
df.describe()
```

```python
# Display information about the DataFrame's data types and memory usage
df.info
```

```python
# bar chart of Unemployment rate
df[' Estimated Unemployment Rate (%)'].value_counts().plot(kind='bar')
```

```python
# bar chart of Employed
df[' Estimated Employed'].value_counts().plot(kind='bar')
```

```python
# Here i showed the  Estimated Unemployment Rate (%) in the different state of india.
df.groupby("Region")[" Estimated Unemployment Rate (%)"].agg(["mean",'min',"max","count"]).plot(kind="bar")
```

```python
# Convert the 'Date' column to datetime format
df[' Date'] = pd.to_datetime(df[' Date'])
```

```python
# Set the 'Date' column as the index
df.set_index(' Date', inplace=True)
```

**Rate of unemployment is increased 5/2020 in the covide 19 year is 77%.
It is higher then other month in the year 2020 .**  

```python
# Visualize the unemployment rate over Date
plt.figure(figsize=(10, 6))
plt.plot(df.index, df[' Estimated Unemployment Rate (%)'], marker='o', linestyle='-')
plt.title(' Unemployment Rate Over date')
plt.xlabel(' Date')
plt.ylabel(' Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
# Group the data by "Date" and calculate various aggregate statistics for the "Estimated Unemployment Rate (%)"
# Plot a bar chart to visualize the aggregate statistics for each region
df.groupby(" Date")[" Estimated Unemployment Rate (%)"].agg(["mean",'min',"max","count"]).plot(kind="bar")
```

Rate of Employment in 2/2020 in the covide 19 year is 60%. It is higher then other month in the year 2020.
after the month the rate of Employed person is constantly Decreasing

```python
# Visualize the unemployment rate over date
plt.figure(figsize=(20,6))
plt.plot(df.index,df[' Estimated Employed'],marker='o',linestyle="-")
plt.title('Employed Rate Over Date')
plt.xlabel(' Date')
plt.ylabel(' Estimated Employed')
plt.grid(True)
plt.show()
```

```python
# Group the data by "Date" and calculate various aggregate statistics for the "Estimated employment "
# Plot a bar chart to visualize the aggregate statistics for each region
df.groupby(" Date")[" Estimated Employed"].agg(["mean",'min',"max","count"]).plot(kind="bar")
```
